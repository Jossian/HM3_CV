```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib as mpl
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("example1_2D.csv", header = None, dtype=np.float64)
data = data.values.tolist()

values = []
x = []
y = []

frontiers = []

while len(data) > 0:
    #print(data)

    notdominated = []
    dominated = []

    for i in range(len(data)):

        dom = 0

        for j in range(len(data)):
            paret = 0
            for k in range(len(data[i])):

                if data[i][k] > data[j][k]:
                    paret += 1
                else:
                    paret -= 1
                if data[i][k] < data[j][k]:
                    paret += 1
                else:
                    paret -= 1
            
            if paret > 0:
                dom += 1
            
        if dom > 0:
            dominated.append(data[i])
        else:
            notdominated.append(data[i])

    data = dominated
    frontiers.append(notdominated)

n_lines = 10
cmap = mpl.colormaps['gist_ncar']
colors = cmap(np.linspace(0, 1, n_lines))

plt.rcParams["axes.prop_cycle"] = plt.cycler("color", colors)
# plt.rcParams["axes.prop_cycle"] = plt.cycler("color", mcolors.XKCD_COLORS)

for frontier in frontiers:
    frontier.sort(key = lambda row: row[1])
    frontier.sort(key = lambda row: row[0], reverse=True)
    x = []
    y = []
    for i in range(len(frontier)):
        x.append(frontier[i][0])
        y.append(frontier[i][1])
        plt.plot(x, y)
plt.show()
```
